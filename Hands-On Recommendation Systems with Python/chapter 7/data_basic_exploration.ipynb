### Movie metadata data preparation and preprocessing

```python
movies_metadata_df.head()
```

```python
def convert_to_int(x):
    try:
        return int(x)
    except:
        return np.NaN
```

```python
movies_metadata_df["movieId"] = movies_metadata_df["id"].apply(convert_to_int)
```

```python
movies_metadata_df.head()
```

```python
movies_metadata_df[movies_metadata_df["movieId"].isna()].shape[0]
```

```python
movies_metadata_df[movies_metadata_df["movieId"].isna()]
```

**There are only 3 rows where id of the movie is in the form of date as shown above. Upon Closer Inspection, we can find that most of the values like original_title, title and overview are not clearly stated. Hence, we drop these rows.**

```python
movies_metadata_df.dropna(subset=["movieId"], inplace=True)
movies_metadata_df.reset_index(drop=True, inplace=True)
```

```python
movies_metadata_df.columns
```

```python
relevant_columns = ["movieId", "original_title", "genres", "original_language", "overview", "production_companies", "production_countries"]
cleaned_metadata_df = movies_metadata_df[relevant_columns]
cleaned_metadata_df.head()
```

```python
cleaned_metadata_df.info()
```

```python
from ast import literal_eval

def convert_obj_to_list(column):
    global cleaned_metadata_df
    cleaned_metadata_df[column] = cleaned_metadata_df[column].fillna('[]')
    cleaned_metadata_df[column] = cleaned_metadata_df[column].apply(literal_eval)
    cleaned_metadata_df[column] = cleaned_metadata_df[column].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
convert_obj_to_list("genres")
cleaned_metadata_df.head()
```

```python
convert_obj_to_list("production_companies")
convert_obj_to_list("production_countries")
cleaned_metadata_df.head()
```

```python
for column in ["genres", "production_companies", "production_countries"]:
    cleaned_metadata_df[column] = cleaned_metadata_df[column].apply(lambda x: " ".join(x))
cleaned_metadata_df.head()
```

```python
cleaned_metadata_df["movieId"] = cleaned_metadata_df["movieId"].astype(int)
cleaned_metadata_df.head()
```

```python
cleaned_metadata_df.to_csv("data/cleaned_movies_metadata.csv", index=False)
```
