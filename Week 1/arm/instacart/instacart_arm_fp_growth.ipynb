**What we need is Order id and its corresponding products bought**

- To obtain this, we have order_id and product_id from **order_products__train.csv**
- Map product_id to product_name from **products.csv**

```python
products_df = pd.read_csv("../../embedding/product_embedding/data/products.csv")
display(products_df.head())

orders_df = pd.read_csv("../../embedding/product_embedding/data/order_products__train.csv")
display(orders_df.head())
```

```python
print(products_df.info())
```

```python
print(orders_df.info())
```

**There are not any null values in our dataframes**

```python
transactions = pd.merge(orders_df, products_df, on="product_id", how="left")
transactions.head()
```

```python
transactions.info()
```

```python
transactions.groupby('order_id')["product_name"].unique()[:2]
```

```python
order_transactions = pd.DataFrame(transactions.groupby('order_id')['product_name'].nunique().index)
order_transactions.head()
```

```python
order_transactions["products"] = transactions.groupby('order_id')["product_name"].unique().values
order_transactions.head()
```

```python
order_transactions.info()
```

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = order_transactions["products"].tolist()
dataset[:2]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset)
te_ary = te_ary.transform(dataset)
type(te_ary), te_ary.shape
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
df.info()
```

```python
frequent_itemsets = fpgrowth(df, min_support=0.0001, use_colnames=True)
type(frequent_itemsets)
```

```python
frequent_itemsets.head()
```

```python
frequent_itemsets.info()
```

```python
frequent_itemsets = frequent_itemsets.sort_values(by="support", ascending=False)
frequent_itemsets.head()
```

```python
frequent_itemsets.head(10)
```

**It shows that banana, Organic bananas, strawberries, spinach are the most commonly bought items.**

```python
for index, row in frequent_itemsets[:100].iterrows():
    if len(tuple(row["itemsets"])) > 1:
        print(index, row["support"], row["itemsets"])
    
```
