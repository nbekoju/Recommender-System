```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
plt.rcParams["figure.figsize"] = [16, 8]

A = imread("photo.jpeg")
X = np.mean(A, -1)

img = plt.imshow(X)
img.set_cmap("gray")
plt.axis("off")
plt.show()
```

```python
type(X), X.shape
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
print(type(U), type(S), type(VT))
print(U.shape, S.shape, VT.shape)
```

```python
S = np.diag(S)
type(S), S.shape
```

```python
j = 0
for r in (5, 20, 100, 200):
    Xapprox = U[:, :r] @ S[:r, :r] @ VT[:r, :]
    plt.figure(j + 1)
    j += 1

    img = plt.imshow(Xapprox)
    img.set_cmap("gray")
    plt.axis("off")
    plt.title("r = " + str(r))
    plt.show()
```

```python
S.shape
```

```python
np.diag(S).shape
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title("Singular Value")
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))
plt.title("Singular Values: Cumulative Sum")
plt.show()
```

**The cumulative sum plot shows that 90% of the energy is obtained from the first 200 columns.**
